```python
import pandas as pd 
```

```python
#reading csv file 
df = pd.read_csv("reviews.csv")
df_2 = pd.read_csv("people.csv")
```

```python
df_2.head()
```

```python
df_2.isnull().sum()
```

```python
#1. What is the average age of an employee at the company?
average_age = df_2['age'].mean()
round(average_age,2)
```

```python
#2. What is the average compensation of employees under 30 years of age (inclusive)?
df_under_equal_30 = df_2[df_2['age'] <= 30]
round(df_under_equal_30['compensation'].mean(),2)
```

```python
#3. What is the 75th percentile for compensation?
import numpy as np
percentile_75 = np.percentile(df_2['compensation'], 75) 
round(percentile_75,2)
```

```python
#4. For countries with two or more employees, what is the average compensation?
df_countries = df_2['country'].value_counts()
df_countries_morethan2= df_countries[df_countries>=2].index
df_morethan2emp=df_2[df_2['country'].isin(df_countries_morethan2)]
round(df_morethan2emp['compensation'].mean(),2)
```

```python
#5. Determine top performing employees
df_2['person_id']=df_2['id'] #creating new column to match person_id in reviews csv 
merged_df = pd.merge(df, df_2, on='person_id') #merging based on person_id
merged_df.sort_values(by='rating', ascending=False).head(10).reset_index()
```
